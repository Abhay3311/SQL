```python
import pandas as pd
import sqlite3
import requests

# URL of the CSV file on GitHub
url = 'https://raw.githubusercontent.com/Abhay3311/SQL/main/Untitled%20spreadsheet%20-%20Sales.csv'

# Download the CSV file
response = requests.get(url)
with open('superstore.csv', 'wb') as file:
    file.write(response.content)

# Load the CSV file into a pandas DataFrame
data = pd.read_csv('superstore.csv')

# Create a SQLite database (or connect to an existing one)
conn = sqlite3.connect('superstore.db')

# Load the DataFrame into the SQLite database
data.to_sql('superstore', conn, if_exists='replace', index=False)
```

Retrive the whole data provided by the Super Store management using SQL query.

```python
pd.read_sql_query("select * from superstore",conn)
```

Retrieve the OrderID and CustomerID columnS from the given dataset.

```python
pd.read_sql_query("select OrderID, CustomerID from superstore",conn)
```

```python
pd.read_sql_query(""" Select
                      OrderID, State,City,Category,ActualSales
                      from superstore
                      Order by ActualSales Desc """,conn)
```

Retrieve all the OrderYear along with their State and ActualSales

Ordered by State in the ascending order.


Ordered by ActualSales in the descending order.

```python
pd.read_sql_query(""" Select
                      OrderYear, State,ActualSales
                      from superstore
                      Order by State Asc,ActualSales Desc
                       """,conn)
```

```python
pd.read_sql_query(""" Select
                      OrderYear, State,ActualSales
                      from superstore
                      Order by State Asc,ActualSales Desc
                      Limit 5
                       """,conn)
```

```python
pd.read_sql_query(""" Select
                      OrderYear, State,ActualSales
                      from superstore
                      Where OrderYear = 2023
                       """,conn)
```

```python
pd.read_sql_query(""" Select
                      OrderYear, State,ActualSales
                      from superstore
                      Where OrderYear = 2023
                      Order By ActualSales
                       """,conn)
```

```python
pd.read_sql_query(""" Select
                      OrderYear, State,ActualSales
                      from superstore
                      Where OrderYear = 2023
                      Order By ActualSales Desc
                      Limit 10
                       """,conn)
```

Double WHERE

```python
pd.read_sql_query(""" Select
                      OrderYear, State,ActualSales
                      from superstore
                      Where OrderYear = 2023
                      and State = 'Florida'
                      Order By ActualSales Desc
                      Limit 10
                       """,conn)
```

```python
pd.read_sql_query(""" Select
                      OrderYear, State,ActualSales
                      from superstore
                      Where OrderYear IN (2022,2023)
                      and ActualSales <= 500
                      and ActualSales >= 300
                      Order By ActualSales Desc
                      Limit 10
                       """,conn)
```

```python
pd.read_sql_query(""" Select
                      OrderYear, State, ActualSales
                      from superstore
                      Where State IN ('California','Texas','New York')
                      and ActualSales >= 600
                      Order By ActualSales Desc
                      Limit 10
                       """,conn)
```
